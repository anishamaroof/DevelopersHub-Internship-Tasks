## 📌 Introduction & Problem Statement

In this task, we explore the classic **Iris Dataset**, which contains measurements of 150 iris flowers from three different species:
- *Iris Setosa*
- *Iris Versicolor*
- *Iris Virginica*

Each flower has 4 features measured:
- Sepal Length (cm)
- Sepal Width (cm)
- Petal Length (cm)
- Petal Width (cm)

**Goal:** Load, inspect, and visualize the dataset to understand patterns and relationships between features.

---

## 📦 Step 1: Import Libraries

```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting plot style for better visuals
sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (8, 5)

print('✅ Libraries imported successfully!')
```

## 📂 Step 2: Load the Dataset

```python
# Loading Iris dataset directly from seaborn (no need to download any CSV)
df = sns.load_dataset('iris')

print('✅ Dataset loaded successfully!')
print(f'Dataset shape: {df.shape}')
```

## 🔍 Step 3: Dataset Understanding & Inspection

```python
# Display first 5 rows
print('🔹 First 5 rows of the dataset:')
df.head()
```

```python
# Check shape of the dataset
print(f'🔹 Shape of dataset: {df.shape}')
print(f'   → {df.shape[0]} rows and {df.shape[1]} columns')
```

```python
# Display all column names
print('🔹 Column names:')
print(df.columns.tolist())
```

```python
# Data types and non-null counts
print('🔹 Dataset Info:')
df.info()
```

```python
# Statistical summary
print('🔹 Statistical Summary:')
df.describe()
```

```python
# Check for missing values
print('🔹 Missing Values per Column:')
print(df.isnull().sum())
print('\n✅ No missing values found — dataset is clean!')
```

```python
# Count of each species
print('🔹 Species Distribution:')
print(df['species'].value_counts())
```

## 📊 Step 4: Exploratory Data Analysis (EDA) — Visualizations

---

### 📈 4.1 Scatter Plot — Relationship Between Variables

```python
# Scatter Plot: Sepal Length vs Sepal Width
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x='sepal_length',
    y='sepal_width',
    hue='species',
    palette='Set1',
    s=80
)
plt.title('Scatter Plot: Sepal Length vs Sepal Width', fontsize=14, fontweight='bold')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
print('📌 Insight: Setosa species has smaller sepal length but wider sepal width compared to others.')
```

```python
# Scatter Plot: Petal Length vs Petal Width
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=df,
    x='petal_length',
    y='petal_width',
    hue='species',
    palette='Set2',
    s=80
)
plt.title('Scatter Plot: Petal Length vs Petal Width', fontsize=14, fontweight='bold')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
print('📌 Insight: Petal dimensions clearly separate all 3 species — great feature for classification!')
```

```python
# Pairplot — All feature relationships at once
print('🔹 Pairplot: All Feature Combinations')
sns.pairplot(df, hue='species', palette='Set1', plot_kws={'alpha': 0.6})
plt.suptitle('Pairplot of All Iris Features', y=1.02, fontsize=14, fontweight='bold')
plt.show()
print('📌 Insight: Petal length & petal width show the clearest separation between species.')
```

### 📊 4.2 Histogram — Data Distribution

```python
# Histograms for all 4 features
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
colors = ['#4C72B0', '#DD8452', '#55A868', '#C44E52']

for i, (feature, color) in enumerate(zip(features, colors)):
    row, col = divmod(i, 2)
    axes[row][col].hist(df[feature], bins=20, color=color, edgecolor='white', alpha=0.85)
    axes[row][col].set_title(f'Distribution of {feature.replace("_", " ").title()}', fontweight='bold')
    axes[row][col].set_xlabel(feature.replace('_', ' ').title() + ' (cm)')
    axes[row][col].set_ylabel('Frequency')

plt.suptitle('Histograms: Distribution of All Features', fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
print('📌 Insight: Sepal length is roughly normally distributed. Petal features show bimodal distribution indicating species differences.')
```

```python
# Histogram with KDE (density curve) per species
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features):
    row, col = divmod(i, 2)
    for species in df['species'].unique():
        subset = df[df['species'] == species][feature]
        axes[row][col].hist(subset, bins=15, alpha=0.5, label=species, edgecolor='white')
    axes[row][col].set_title(f'{feature.replace("_", " ").title()} by Species', fontweight='bold')
    axes[row][col].set_xlabel('cm')
    axes[row][col].set_ylabel('Frequency')
    axes[row][col].legend()

plt.suptitle('Feature Distribution by Species', fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
print('📌 Insight: Each species shows distinct distribution patterns, especially in petal measurements.')
```

### 📦 4.3 Box Plot — Outliers & Spread

```python
# Box plots for all features grouped by species
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features):
    row, col = divmod(i, 2)
    sns.boxplot(
        data=df,
        x='species',
        y=feature,
        palette='Set2',
        ax=axes[row][col]
    )
    axes[row][col].set_title(f'Box Plot: {feature.replace("_", " ").title()} by Species', fontweight='bold')
    axes[row][col].set_xlabel('Species')
    axes[row][col].set_ylabel(feature.replace('_', ' ').title() + ' (cm)')

plt.suptitle('Box Plots: Feature Spread & Outliers by Species', fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()
print('📌 Insight: Virginica has the largest petal dimensions. A few outliers visible in sepal width for Setosa.')
```

### 🌡️ 4.4 Bonus: Correlation Heatmap

```python
# Heatmap to show correlation between features
plt.figure(figsize=(7, 5))
numeric_df = df.drop(columns=['species'])
correlation = numeric_df.corr()

sns.heatmap(
    correlation,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    linewidths=0.5,
    square=True
)
plt.title('Correlation Heatmap of Iris Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
print('📌 Insight: Petal length and petal width are highly correlated (0.96). Sepal width has weak/negative correlation with other features.')
```
